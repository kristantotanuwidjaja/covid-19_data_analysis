```python
labels = ['India', 'Indonesia', 'Philippines', 'Burma', 'Malaysia']
total_cases = [total_in, total_id, total_ph, total_bu, total_my]
total_popu = [1380004385, 273523615, 109581078, 51419420, 28334000]

x = np.arange(len(labels))  # the label locations
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, total_cases, width, label='Confirmation')
rects2 = ax.bar(x + width/2, total_popu, width, label='Population')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Confirmation and Population')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

## Monthly Cases in Indonesia

```python
unique_year = pd.DatetimeIndex(se_asia_confirm['Date']).year.unique()
unique_month = pd.DatetimeIndex(se_asia_confirm['Date']).month.unique()
list_sum = []
region = 'Indonesia'

d = pd.DatetimeIndex(se_asia_confirm['Date'])
for y in unique_year :
    for m in unique_month :
        tmp = se_asia_confirm[(d.month == m) & (d.year == y)]
        print_date = str(y) + '/' + str(m)
        list_sum.append([print_date, tmp[region].sum()])

list_day = [i[0] for i in list_sum]
list_sum = [i[1] for i in list_sum]
preview = pd.DataFrame(list_day, columns=['Date'])
preview['Total Cases per Month'] = list_sum

print('Region : ' + region)
preview
```

# Time Series using ahlive

## Read Data

```python
raw_confirm_cases=pd.read_csv("./Datasets/RAW_global_confirmed_cases.csv")
raw_confirm_cases
```

```python
df_tidy = raw_confirm_cases.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date', value_name='Cases')
display(df_tidy)
```

```python
countries = ['Indonesia', 'Malaysia', 'Philippines', 'Singapore', 'Thailand']
df_subset = df_tidy.loc[df_tidy['Country/Region'].isin(countries)]
display(df_subset)
```

```python
df_countries = df_subset.groupby(['Date', 'Country/Region'])[['Cases']].sum().reset_index()
df_countries = df_countries.loc[df_countries['Date'] >= '5/01/21']
display(df_countries)
```

```python
ah_df = ah.DataFrame(df_countries, 'Date', 'Cases')
ah_df
```

```python
ah_df = ah.DataFrame(df_countries, 'Date', 'Cases', label='Country/Region')
ah_df
```

```python
df_dts = df_countries.copy()
df_dts['Date'] = pd.to_datetime(df_dts['Date'])
df_dts = df_dts.loc[df_dts['Date'] >= '7/1/20']
df_dts = df_dts.groupby([
    pd.Grouper(key='Date', freq='7D'), 'Country/Region'
])['Cases'].last().reset_index()
df_scale = df_dts.sort_values('Date')
df_scale['Cases'] /= 1e6
```

```python
ah_df = ah.DataFrame(
    df_scale, 'Date', 'Cases', label='Country/Region',
    figsize=(10, 7), animate='test'
).config('legend', sortby='y')
ah_df.render()
```

```python
ah_df = ah.DataFrame(
    df_scale, 'Date', 'Cases', label='Country/Region',
    state_labels='Date', inline_labels='Cases',
    figsize=(10, 7), ylabel='Confirmed Cases [million]',
    animate='test'
).config('yticks', format='%.0f').config('legend', sortby='y')
ah_df.render()
```
